````python
import numpy as np
from matplotlib import pyplot as plt
````

# Equation

```python
y = m*x + c # 2.18 + 2.227*x
```

````python
# Data
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

plt.scatter(x, y)
````

````python
m_cur = 0.0
b_cur = 0.0
learning_rate = 0.02
n = x.size
iterations = 200
cost = []

for i in range(iterations):
    y_pred = m_cur*x + b_cur
    
    _cost = (1/n)*sum([val**2 for val in (y-y_pred)])
    cost.append(_cost)

    _m = -(2/n)*sum(x*(y-y_pred))
    _b = -(2/n)*sum(y-y_pred)

    m_cur -= learning_rate * _m
    b_cur -= learning_rate * _b
    plt.plot(x, y_pred)
    print(f"m {m_cur}, b {b_cur}, iteration {i}")
````

## Check the accuracy of model i.e, `Cofficient of determination`

````python
from sklearn.metrics import r2_score

r2 = r2_score(y, y_pred)
r2
````

# Cost function variation with iteration 

````python
index = np.arange(iterations)
plt.plot(index, cost)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
````
