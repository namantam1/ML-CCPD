```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("advertising.csv")
df.describe()
```

```python
df.head()
```

```python
X = df[["TV", "Radio", "Newspaper"]]
y = df["Sales"]
```

```python
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```
