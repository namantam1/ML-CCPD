```python
import numpy as np
# import cv2
import matplotlib.pyplot as plt
```

```python
# sharpen
flt = {
  "sharpen": np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
  ]),
  "blur": np.array([
    [0.0625, 0.125, 0.0625],
    [0.125, 0.25, 0.125],
    [0.0625, 0.125, 0.0625],
  ]),
  "bottom_sobel": np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
  ]),
  "emboss": np.array([
    [-2, -1, 0],
    [-1, 1, 1],
    [0, 1, 2]
  ]),
  "identity": np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
  ]),
  "left_sobel": np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
  ]),
  "outline": np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
  ]),
  "right_sobel": np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
  ]),
  "top_sobel": np.array([
    [1, 2, 1],
    [0, 0, 0],
    [-1, -2, -1]
  ])
}

def convolute(img ,f):
  final_img = np.zeros(shape=(img.shape[0]-2, img.shape[1]-2))

  for i in range(final_img.shape[0]):
    for j in range(final_img.shape[1]):
      final_img[i][j] = np.sum(flt[f] * img[i:i+3, j:j+3])
      if (final_img[i][j] < 0):
        final_img[i][j] = 0
      if (final_img[i][j] > 255):
        final_img[i][j] = 255
  return final_img

def show(img):
    plt.imshow(img, cmap="gray")
```

```python
img = plt.imread("taj.jpg").sum(axis=2)/3
```

```python
show(convolute(img, "blur"))
```
